```python

data = {
    "Nombre de nœuds": [25, 50, 75, 80 , 90 , 100],
    "Temps d'exécution (s)": [0.0208, 0.5490, 8.1325, 36.5404 , 200.4884, 4291.9065],
    "Nombre d'itérations": [1521, 27910, 234222 , 1041347, 5220844 ,90554621 ]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Nombre de nœuds"], df["Temps d'exécution (s)"], marker='o', linestyle='-')
plt.xlabel("Nombre de nœuds")
plt.ylabel("Temps d'exécution (s)")
plt.title("Temps d'exécution en fonction du nombre de nœuds")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Nombre de nœuds"], df["Nombre d'itérations"], marker='o', linestyle='-')
plt.xlabel("Nombre de nœuds")
plt.ylabel("Nombre d'itérations")
plt.title("Performance en fonction du nombre de nœuds")
plt.grid()
plt.show()
```
